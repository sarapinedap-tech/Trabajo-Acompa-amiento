Exploración de datos medicamentos de venta libre.

En este notebook se realiza la exploración inicial del conjunto de datos obtenido desde Datos Colombia.
Se cargan los datos, se revisan las columnas disponibles y se verifica la cantidad de los mismos.

```python
import requests
import pandas as pd
```

```python
base_url = "https://www.datos.gov.co/resource/ec4u-mzse.json"
```

```python
datos = []

for offset in range(0, 10000, 1000):
    params = {
        "$limit": 1000,
        "$offset": offset
    }
    response = requests.get(base_url, params=params)
    datos.extend(response.json())
```

```python
df = pd.DataFrame(datos)
df.head()
```
